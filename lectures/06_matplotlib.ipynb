# Chapter 06: Data Visualization with matplotlib

One of the most fundamental tasks in data analysis is to visualize the data. This can help during the exploratory phase of data analysis, and it can also help you to understand any metrics extracted from that data. Examples of such processes include identification of outliers, identifying patterns in the data, and understanding the distribution of the data.

The most popular visualization tool in Python is matplotlib. It is a Python 2D plotting library that produces publication quality figures in a variety of hardcopy formats and interactive environments across platforms. It can be used in Python scripts, the Python and IPython shells, web application servers, and various graphical user interface toolkits.

Plot the same using np.linspace() instead of np.arange() and with double marker frequency:

`numpy.arange()` and `numpy.linspace()` are both functions in the NumPy library used to create arrays with evenly spaced values, but they differ in how you define the spacing and the inclusion of the endpoint.

`arange()`:

The start value is included, but the end value is excluded. The step size is explicitly provided by the user.  Ideal when you need to control the increment between values and the number of elements is determined by the step size.

`linspace()`:

The start and end values are inclusive by default, and the user provides the number of samples in between. Best when you need a specific number of elements in your array and the exact start and end points are crucial.

```python
np.linspace(0,20,400)
```

```python
x = np.linspace(0,20,40)
y = x
z = x ** 2
plt.plot(x, y)
plt.plot(x, z, color='g', linestyle='dashed', marker='o')
plt.title("demo plot")
plt.xlabel("input x")
plt.ylabel("output x and x^2")
plt.legend(["x","$x^{2}$"])
#plt.xlim(5,15)
plt.show()
```

#### Example using builtin trigonometric functions:

```python
#x = np.arange(-2 * np.pi, 2 * np.pi, 0.1)
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = np.sin(x)
z = np.cos(x)
plt.plot(x, y, label = 'sin(x)')
plt.plot(x, z, label = 'cos(x)')
plt.xlabel('x')
plt.ylabel('sin(x) and cos(x)')
plt.title('sine and cosine in+/-2pi')
plt.grid()
plt.legend()
# plt.legend(['sin(x)', 'cos(x)'])
plt.show()
```

```python
x = np.arange(-2 * np.pi, 2 * np.pi, 0.1)
y = np.sin(x)
z = np.cos(x)

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(7.5,1.5))
ax[0].grid(True, which='both')
ax[0].axhline(y=0, color='k')
ax[0].axvline(x=0, color='k')
ax[1].grid(True, which='both')
ax[1].axhline(y=0, color='k')
ax[1].axvline(x=0, color='k')

ax[0].plot(x, y, label = 'sin(x)', color="red")
ax[1].plot(x, z, label = 'cos(x)', color="green")

ax[0].set_title('sin(x)', color="red")
ax[1].set_title('cos(x)', color="green")

ax[0].set_xlabel('x')
ax[1].set_xlabel('x')

ax[1].set_ylabel('sin(x) and cos(x)')

ax[0].legend()
ax[1].legend()

plt.show()
```

```python
x = np.arange(-2 * np.pi, 2 * np.pi, 0.1)
y = np.sin(x)
z = np.cos(x)

fig, ax = plt.subplots(nrows=2,ncols=2, figsize=(12.5,6))

ax[0,0].plot(x, y, label = 'sin(x)', color="red")
ax[0,1].plot(x, z, label = 'cos(x)', color="green")
ax[1,0].plot(x, y + z, label = 'sin(x) + cos(x)', color="pink")
ax[1,1].plot(x, z - y, label = 'cos(x) - sin(x)', color="purple")

ax[0,1].axhline(y=0, color='k')
ax[0,1].axvline(x=0, color='k')
ax[1,1].grid(True, which='both')
ax[1,1].axhline(y=0, color='k')
ax[1,1].axvline(x=0, color='k')
ax[0,0].legend()
ax[0,1].legend()
ax[1,0].legend()
ax[1,1].legend()

plt.show()
```

#### A histogram is a type of (bar) graph that represents the distribution of numerical data. It consists of adjacent bins (bars or rectangles) where each rectangle’s height is representing the frequency of data points in that bin’s range. The width of each bin represents a specific interval of values.

Histograms are useful for visualizing the distributions of the data, identifying its shape and patterns such as skewness, modality, and detecting outliers.

#### A stackplots is a graph showing how different components contribute to the whole.
It is useful to display the distribution of data where each part is stacked on top of the others, showing a cumulative effect.

#### Example for stack plot:

```python
o = range(1, len(x)+1)
plt.stackplot(o,x,y1,y2, colors=['r','g','b'])
plt.plot(o,x, color='k')
plt.grid()
plt.show()
```

### The Dark Side of Pie Charts: Cognitive Challenges and Spatial Inefficiency
- Poor Data Communication: Pie charts are not effective at conveying data.
- Cognitive Limitations: The human brain struggles with comparing angles, lacks a scale, and finds it difficult to read accurate values.
- Spatial Drawbacks: They occupy more space and are harder to read than other chart types.
- Data Scientists hate them for the above reasons and also just because they can.

#### Load and plot data from file:

Let us generate some random data and plot it:
1. Generate 100 rows of random data with x, y, z columns
2. Save this data to 'testload.csv' file



```python
import numpy as np

# Generate 100 random data points for x, y, z
np.random.seed(42)                      # For reproducible results
x = np.random.randint(1, 101, 150)      # Random integers between 1-100
y = np.random.randint(1, 201, 150)
z = np.random.randint(1, 101, 150)

# Stack the arrays and transpose to get 100 rows x 3 columns
data = np.column_stack((sorted(x), sorted(y), sorted(z)))

# Save to CSV file
np.savetxt('testload.csv', data, delimiter=',', fmt='%d')

print(f"Generated {len(data)} rows of x, y, z data and saved to 'testload.csv'")
print("First 5 rows:")
print(data[:5])
```

```python
# loading will error out if csv was not provided
x, y, z = np.loadtxt('testload.csv', delimiter=',', unpack=True)
plt.plot(x, y, color="red")
plt.plot(x, z)
plt.show()
```
